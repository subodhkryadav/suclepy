```python
from suclepy import auto_clean
```

```python
# Load example dirty CSV
data_path = "../data/example_dirty.csv"
```

```python
# Run auto_clean
df_clean, roles, validation_report = auto_clean(data_path)
```

```python
# Display cleaned data
print("Cleaned DataFrame:")
print(df_clean.head())
```

```python
# Display inferred roles
print("\nColumn Roles:")
for col, role in roles.items():
    print(f"{col}: {role}")
```

```python
# Display validation report
print("\nValidation Report:")
for key, value in validation_report.items():
    print(f"{key}: {value}")
```
